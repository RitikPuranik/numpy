DEEP AND SHALLOW COPY

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice
```

```python
slice = slice * 10
slice
```

```python
#no changes are made in original array but if there was an list 
#then the changes would have been made in original list as well 
# because of mutability of list but here we are working with array 
# which is immutable so no changes are made in original array.

# in respective of array  it is called as shallow copy
# in respective of list it is called as deep copy

a
```

```python
# this is an example of deep copy of array
a
b = a
```

```python
b[0] = 99
b
```

```python
a
```

```python
# alternative for it is to use copy method of array
a
b = a.copy()
b[0] = 99
b
```

STACKING ARRAY

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.stack((a,b))
```

```python
# to stack two arrays vertically we can use vstack method of numpy
np.vstack((a,b))
```

```python
# to stack two arrays horizontally we can use hstack method of numpy
np.hstack((a,b))
```

```python
# to stack two arrays column wise we can use column_stack method of numpy
np.column_stack((a,b))
```

SPLITTING ARRAY

```python
c = np.arange(16).reshape(4,4)
c
```

```python
# to split an array into multiple sub-arrays horizontally 
# we can use hsplit method of numpy
np.hsplit(c, 2)
```

```python
# to split an array into multiple sub-arrays vertically 
# we can use vsplit method of numpy
a = np.vsplit(c, 4)
a
```

```python
#the array does not split in 1d array but 
# it splits in 2d array
for i in a:
    print(i)
```
